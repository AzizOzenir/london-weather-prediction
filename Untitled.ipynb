```python
df.dropna(subset=["cloud_cover", "sunshine","global_radiation","max_temp","mean_temp","min_temp","precipitation","pressure"], inplace=True)
mean_value = df['snow_depth'].mean()
df['snow_depth'].fillna(mean_value, inplace=True)
df.isnull().sum()
```

```python
def effectOnWeather(var):
    if var not in ["date","min_temp","mean_temp", "max_temp"]:
        plt.figure(figsize=(8, 6))
        plt.subplot(1, 1, 1)
        sns.barplot(x=df[var], y=df["mean_temp"], palette='plasma')
        plt.title(f' {var} - VS - Mean Temperature')  # Set a title
        plt.xlabel(var)
        plt.ylabel('Mean Temperature')
        
for column in df.columns:
        effectOnWeather(column)
        plt.show()

plt.tight_layout()  # Adjust subplot layout for a clean presentation
plt.show()
```

# MODEL

```python
df["date"] = df["date"].astype(str)
print(df.dtypes)
```

```python
df["date"]=df["date"].str[4:6]
df.head()
```

```python
df = pd.get_dummies(df,columns=["date"],prefix="month")
df.head()
```

```python
y=df["mean_temp"]
x=df.drop(["mean_temp","max_temp","min_temp"],axis=1)

print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.metrics import r2_score,mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from xgboost import XGBRegressor
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.15,random_state=42)
print("x_train",x_train.shape)
print("x_test",x_test.shape)
print("y_train",y_train.shape)
print("y_test",y_test.shape)
```

```python
LR = LinearRegression()
XGBR = XGBRegressor()
SVR = SVR()
```

```python
models = [LR,XGBR,SVR]
for i in models:
    print(i)
    i.fit(x_train,y_train)
    print("test")
    print(i,"score is",r2_score(y_test,i.predict(x_test)),mean_absolute_error(y_test,i.predict(x_test)))
    
    print("train")
    print(i,"score is",r2_score(y_train,i.predict(x_train)),mean_absolute_error(y_train,i.predict(x_train)))
    
    print("-------------------  ")
    print("-------------------  ")
    print("-------------------  ")
```
